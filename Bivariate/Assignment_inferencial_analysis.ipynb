 # Z_Score:  Conversion of Normal Ditribution to Standard Normal Distribution....

```python
def STD_Normal_Distribution(preprocessed_ds):
    import seaborn as sns
    from scipy.stats import zscore
    
    df=pd.DataFrame(preprocessed_ds)
    
    # Calculate the Zscore for each column and plot that points using histplot and kdeplot fns
    z_scores_df=df.apply(zscore)
    #sns.distplot(z_scores_df,kde=True)
    
    sns.histplot(data=z_scores_df, color='green',stat='density')
    sns.kdeplot(data=z_scores_df, color='red')
    
    #Display the Normal Distribution and STD Normal Distribution
    print("Original Normal Distribution:")
    print(df)
    print("STD Normal Distribution:")
    print(z_scores_df)
    
```

```python
STD_Normal_Distribution(preprocessed_ds["salary"])
```

## 10. What is the Probability Density Function(pdf) of the salary range from 700000 to 900000?

```python
def pdf(preprocessed_ds,start_range,end_range):
    from scipy.stats import norm

    #start_range=70000
    #end_range=80000
    
    mean=preprocessed_ds["salary"].mean()
    std=preprocessed_ds["salary"].std()
    
    # calculate normal distribution using Mean and STD
    dist=norm(mean,std)
    
    # calculate the pdf using norm,pdf() from scipy.stats
    
    values=[]
    for value in range(start_range,end_range):
        values.append(value)
        
    probabilities=[]
    for value in values:
        probabilities.append(dist.pdf(value))
        
                             
    # sum the probabilites to find the area under the curve between the range
        prob=sum(probabilities)
    
    print("Area between range {},{}:{} ".format(start_range,end_range,prob))
```

```python
pdf(preprocessed_ds,700000,900000)
```

# 11. Test the similarity between the degree_t(Sci&Tech)with respect to etest_p and mba_p at significance level of 5%.(Make decision using Hypothesis Testing)

```python
### same group (degree_t(sci&tech))) diff condition (etest_p, mba_p)   - paired TTest
## Is etest_p and mba_p in Sci&Tech are same??? 
## TTest

from scipy.stats import ttest_ind

etest_p=preprocessed_ds[preprocessed_ds["degree_t"]=="Sci&Tech"]["etest_p"]
mba_p=  preprocessed_ds[preprocessed_ds["degree_t"]=="Sci&Tech"]["mba_p"]

ttest_ind(etest_p,mba_p)
```

```python
# Hypothesis Testing:
# H0- Null Hypothesis: There is NO significance similarity between etest_p and mba_p in Sci&Tech
# H1- Altertnate Hypothesis. There is significance similarity between etest_p and mba_p in Sci&Tech

# Conclusion:
# TtestResult(statistic=np.float64(4.532000225151251), pvalue=np.float64(1.4289217003775638e-05), df=np.float64(116.0))

# pvalue=np.float64(1.4289217003775638e-05), pvalue<.05. So Reject Null hypothesis and Accept Alternate hypothesis. 
# so there is significance similarity between etest_p and mba_p in Sci&Tech
```
